```python
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
diag_proc_df = [col for col in data if 'diagnosis' in col.lower() or 'procedure' in col.lower()]
diag_proc_df = data[diag_proc_df]
print(diag_proc_df)
```

```python
diag_prim_proc_df = diag_proc_df.drop(columns = ['diagnosis_code_version','procedure_code_modifier_1','procedure_code_modifier_2'])
print(diag_prim_proc_df)
```

Data Transformation

```python
# Transaction Encoding
patient_transactions = []
for i in range(0,100):
    patient_transactions.append([str(diag_prim_proc_df.values[i,j]) for j in range(0,5)])
print(patient_transactions)
```

```python
from collections import Counter

# Clean the Transactions
cleaned_transactions = [[item for item in transaction if item != 'nan'] for transaction in patient_transactions]

# Calculate the frequency of each item
all_items = [item for sublist in cleaned_transactions for item in sublist]
item_counts = Counter(all_items)
```

```python
# Convert the counts to a DataFrame for easier analysis
item_counts_df = pd.DataFrame(item_counts.items(), columns=['Item', 'Count'])

# Sort the items by frequency
item_counts_df = item_counts_df.sort_values(by='Count', ascending=False)
print(item_counts_df)
```

```python


# Calculate total number of transactions
total_transactions = len(cleaned_transactions)

# Choose the 10th percentile as the minimum support threshold
min_support_count = item_counts_df['Count'].quantile(0.10)
min_support = min_support_count / total_transactions

print(f'Minimum Support: {min_support}')
```

Training Apriori Model

```python
from apyori import apriori
rules = apriori(transactions= patient_transactions, min_support = 0.01, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 5)
```

Visualising the Result

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, support, confidence, lifts))

ResultDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lifts' ])
ResultDataFrame
```

```python
ResultDataFrame.nlargest(n = 100, columns= "Lifts")
```
